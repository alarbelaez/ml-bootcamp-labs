```python
# Para predicciones nuevas:
#  Creamos un dataframe con valores que son 'nuevos'
# mean	12.259259	2789.451525	207.714697	528.023673	1596.137863	777.213061	0.456551	5.178157	0.599766	24.933094

# Estos son datos que van a ser nuevos para nuestro modelo
diccionario_datos = {
    'Well_ID': 1.0,
    'Pressure': 2750.0,
    'Temperature': 200.5,
    'Flow_Rate': 515.1,
    'Pump_Speed': 1600.1,
    'Gas_Oil_Ratio': 999.9,
    'Water_Cut': 0.5,
    'Vibration': 5.12,
    'Maintenance_Required': 0,
    'Downtime': 22.5,
}

new_data = pd.DataFrame(diccionario_datos,index=[0])
```

```python
new_data.columns
```

```python
# Recordamos que nuestro model ya fue importado desde sesiones anteriores y ahora el objetivo es solamente utilizarlo:

model.predict(new_data)
```

```python
# Función que se encarga de automatizar la carga y prediccion de datos nuevos:

def prediccion_datos(well_id, pressure, temp, flow, pump, gas_oil, water, vibration, maintenance, downtime):
    diccionario_datos = {
        'Well_ID': well_id,
        'Pressure': pressure,
        'Temperature': temp,
        'Flow_Rate': flow,
        'Pump_Speed': pump,
        'Gas_Oil_Ratio': gas_oil,
        'Water_Cut': water,
        'Vibration': vibration,
        'Maintenance_Required': maintenance,
        'Downtime': downtime,
    }
    new_data = pd.DataFrame(diccionario_datos,index=[0])
    return model.predict(new_data)
    
```
